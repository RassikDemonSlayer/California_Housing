**np.ceil, np.where** 

Получение дискретных категорий и объединение всех категорий больше 5 в категорию 5

````python
np.ceil(housing['median_income'] / 1.5).value_counts()
````

````python
housing['income_cat'] = np.ceil(housing['median_income'] / 1.5)
````

````python
housing['income_cat'] = housing['income_cat'].where(housing['income_cat']<5, 5.0)
````

````python
housing["income_cat"].value_counts()
````

````python
housing["income_cat"].hist()
````

**sklearn.model_selection StratifiedShuffleSplit**

Стратифицированная выборка на основе категорий дохода

````python
from sklearn.model_selection import StratifiedShuffleSplit
````

````python
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
````

````python
for train_index, test_index in split.split(housing, housing['income_cat']):
    strat_train_set = housing.loc[train_index]
    strat_test_set = housing.loc[test_index]
````

````python
# Доля категорий дохода
housing['income_cat'].value_counts()/len(housing)
````

````python
strat_train_set.drop('income_cat', axis=1, inplace=True)
strat_test_set.drop('income_cat', axis=1, inplace=True)
````
